### Monte Carlo simulation for approximation of pi number
The number π is a mathematical constant, approximately equal to 3.14159. It is defined in Euclidean geometry as the ratio of a circle's circumference to its diameter, and also has various equivalent definitions. The number appears in many formulas in all areas of mathematics and physics. The earliest known use of the Greek letter π to represent the ratio of a circle's circumference to its diameter was by Welsh mathematician William Jones in 1706. It is also referred to as Archimedes's constant.

The simulation works on the principle of generating random points in the square of the page r = 1.
Based on the distance of the point of the center of the coordinate system, it is possible to determine whether the point is inside or outside imaginary quarters of a circle. How is a quarter of the surface of a circle <img src="https://render.githubusercontent.com/render/math?math=\frac{r^2\pi}{4}"> and area of square <img src="https://render.githubusercontent.com/render/math?math={r^2}"> number π can be approximated as  <img src="https://render.githubusercontent.com/render/math?math=\frac{k}{m}*4"> where k is the number of points within a quarter of a circle and m is the total number of simulated random points. As the number of randomly generated points increases, it is possible to calculate the number π with greater precision at the expense of time efficiency.

```python
import random
def pi_aproximation(num_of_spots):

    
    num_of_spots_inside_of_circle = 0
    for i in range(num_of_spots):
        x, y = random.random(), random.random()
        if ((x**2+y**2)**0.5) <= 1:
            num_of_spots_inside_of_circle +=1
        
    aproksimacija_pi = (num_of_spots_inside_of_circle/num_of_spots) * 4
    
    return aproksimacija_pi
```

Now we will calculate the approximated pi number for [10, 100, 10000, 1000000, 1000000000] spots and calculate time needed for task operation.

```python
import time
for i in [10, 100, 10000, 1000000, 1000000000]:
    start = time.monotonic()
    print('pi_approximated:', pi_aproximation(i))
    end = time.monotonic()
    print(f'time needed for calculation of pi number based on {i} spots:', round(end-start,4), 'seconds')
    print(40*'==')
```

```python
def pi_aproximation_za_plotovanje(num_of_spots):
   
    
    inside_of_circle = []
    outside_of_circle = []
    
    num_of_spots_inside_of_circle = 0
    for i in range(num_of_spots):
        x, y = random.random(), random.random()
        if ((x**2+y**2)**0.5) <= 1:
            inside_of_circle.append((x,y))
            num_of_spots_inside_of_circle += 1
        else:
            outside_of_circle.append((x,y))
    
    aproksimacija_pi = (num_of_spots_inside_of_circle/num_of_spots) * 4
    
    return aproksimacija_pi, inside_of_circle, outside_of_circle
```

The beauty of the Data Science is that when you understand the concept, you could easily plot it. With the less spots, the circle is less clear, so the pi number is harder to approximate. As we increase the number of spots, the circle is much clearer, so the pi approximation is better. 

```python
import matplotlib.pyplot as plt
for i in [1000, 10000, 100000]:
    aproksimacija_pi, unutar_kruga, van_kruga = pi_aproximation_za_plotovanje(i)
    plt.scatter([x[0] for x in unutar_kruga], [y[1] for y in unutar_kruga], label='inside_of_circle', color='r', s=0.3)
    plt.scatter([x[0] for x in van_kruga],[y[1] for y in van_kruga], label='outside_of_circle', color='b', s=0.3)
    plt.legend(loc='best')
    plt.show()
```
